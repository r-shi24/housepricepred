```python
#CONVERTING NON NUMERIC TO NUMERIC DATA
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
data['city'] = lb.fit_transform(data['city'])
data['statezip'] = lb.fit_transform(data['statezip'])
data['street'] = lb.fit_transform(data['street'])
```

```python
#SPLITTING THE DATA INTO TARGET AND FEATURE
y = data['price']
x = data.drop('price',axis = 1)
x
from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler()
processed_x = scaler.fit_transform(x)
# processed_x = pd.DataFrame(processed_x)
# processed_x
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 10) #taking any 15 relevant features out of 31 features 
x_pca = pca.fit_transform(x)
x_pca = pd.DataFrame(x_pca)
x_pca
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(processed_x,y,test_size = 0.2, random_state = 42)
```
